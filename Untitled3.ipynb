```python
import pandas as pd
df = pd.read_csv("Drug-Sales-Symptoms.csv")
df.head()
```

```python
# Number of rows and columns
print("Rows and Columns:", df.shape)

# Data types and missing values
df.info()

# See a few rows of data
df.head()
```

```python
# Show total rows and columns
df.shape

# Column names
df.columns

# Data types and missing values
df.info()

# Summary of numbers
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Barplot: Drug sales count
plt.figure(figsize=(8,5))
sns.barplot(x="Drug Name", y="Units Sold", data=df)
plt.title("Units Sold by Drug")
plt.xticks(rotation=45)
plt.show()
```

```python
# Countplot: Symptoms frequency
plt.figure(figsize=(8,5))
sns.countplot(x="Symptom", data=df)
plt.title("Symptom Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot: Units sold
sns.boxplot(x=df["Units Sold"])
plt.title("Units Sold - Boxplot")
plt.show()
```

```python
# Correlation matrix (only for numeric columns like Units Sold)
correlation = df.corr(numeric_only=True)
print(correlation)
```

```python
# Check skewness of 'Units Sold'
print("Skewness of Units Sold:", df['Units Sold'].skew())
```

```python
# Group by Drug and calculate total units sold
df.groupby("Drug Name")["Units Sold"].sum().sort_values(ascending=False)
```

```python
# Group by Symptom
df.groupby("Symptom")["Units Sold"].sum().sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot
plt.figure(figsize=(6,3))
sns.boxplot(x=df["Units Sold"])
plt.title("Boxplot - Units Sold")
plt.tight_layout()
plt.show()
```

- Dolo 650 and Paracetamol are top-selling drugs.
- Fever and headache are common symptoms with high sales.
- No major outliers seen in the boxplot.
- Sales vary across store types: pharmacies sell more than supermarts.
