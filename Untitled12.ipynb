```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Mall_Customers.csv")

# Show first 5 rows (like peeking into your notebook)
print(df.head())

```

```python
X = df.iloc[:, [3, 4]].values  # Income & Spending Score
```

```python
from sklearn.cluster import KMeans

wcss = []  # wcss = error in grouping

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Error)')
plt.grid(True)
plt.show()
```

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 2: Create a dummy dataset (just like Mall_Customers)
np.random.seed(42)
data = {
    'annual_income': np.random.randint(15, 140, 200),
    'spending_score': np.random.randint(1, 100, 200)
}
df = pd.DataFrame(data)

# Step 3: Apply K-Means with k = 5
X = df[['annual_income', 'spending_score']]
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 4: Plot results
plt.figure(figsize=(8, 6))
plt.scatter(X['annual_income'], X['spending_score'], c=y_kmeans, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')
plt.title("K-Means Clustering (Simulated Data, k = 5)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Silhouette Score (higher is better, ranges from -1 to +1)
score = silhouette_score(X, y_kmeans)
print("Silhouette Score:", score)

# Inertia (already seen during Elbow, but repeating for report)
print("Inertia (WCSS):", kmeans.inertia_)
```
